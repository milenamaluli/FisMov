```python
def EqDif(Solucoes, t):
    Theta=Solucoes[0]
    W=Solucoes[1]
    dThetadt = W
    dWdt = -g*math.sin(Theta)/r
    
    return [dThetadt, dWdt]
```

```python
S0=[Theta,0]
```

```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline

Solucao= odeint(EqDif, S0, t)
```

```python
x=[]
y=[]
T=[]
A=[]
for i in range(0,len(Solucao[:,0])):
    Theta=(Solucao[:,0])[i]
    W=(Solucao[:,1])[i]
    x.append(r*math.cos(Theta))
    y.append(r*math.sin(Theta))
    T.append(m*(W**2)*r-m*g*math.cos(Theta))
    A.append(Theta)
```

```python
plt.plot(x, y)
plt.grid(True)
plt.axis('equal')
plt.xlabel('Distância(m)')
plt.ylabel('Altura (m)')
plt.title('Pêndulo com ângulo grande')
plt.show()

plt.plot(t, T)
plt.grid(True)
plt.title('Pêndulo com ângulo grande')
plt.xlabel('Tempo (s)')
plt.ylabel('Tração (N)')
plt.show()
```
